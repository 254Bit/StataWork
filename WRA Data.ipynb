```python
import numpy as np
from scipy import stats
import pandas as pd
df = pd.read_csv('WRA Raw Data10042024.csv')
df
```

```python
df.info()
```

```python
#Rename variables
# Risk_Score (%)
# Sign_Score (%)
# Screening_Score
# Post_Total score (%)
df = df.rename(columns = {'Risk_Score (%)': 'RiskScore','Sign_Score (%)':'SignScore','Screening_Score (%)': 'ScreeningScore','Post_Total score (%)':'PostTotalScore'})
```

```python
df.head()
```

```python
# Convert string object (percentages) to float
df['RiskScore'] = df['RiskScore'].str.rstrip('%').astype(float)
df['SignScore'] = df['SignScore'].str.rstrip('%').astype(float)
df['ScreeningScore'] = df['ScreeningScore'].str.rstrip('%').astype(float)
df['PostTotalScore'] = df['PostTotalScore'].str.rstrip('%').astype(float)

```

```python
df.info()
```

```python
df.head()
```

```python
#Determine the mean, median and p-value of the mentioned variables using arm variable
df1 = pd.crosstab(df['Arm'], df['RiskScore']).mean()
print(df1)
```
